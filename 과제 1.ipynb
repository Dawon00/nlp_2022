```python
from konlpy.tag import Kkma, Okt
import nltk
from collections import Counter

import re
compile = re.compile("[^ ㄱ-ㅣ가-힣]+")

infile = open("소나기.txt", encoding='utf-8')
data=infile.read()

#print(data)   #전체 텍스트

data = compile.sub("", data)
#print(data)   #특수문자 지운 텍스트

#kkma = Kkma()
okt = Okt()
words=okt.nouns(data) #명사만 추출

#print(words)

vocab=Counter(words)
#print(vocab) #빈도 수와 함께 출력

for key,value in vocab.items():
    if value >= 5:
        print(key, value)
```
